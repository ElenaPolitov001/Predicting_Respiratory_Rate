```python
import pandas as pd
df = pd.read_csv('person_12.csv',index_col=0)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
plt.figure(figsize=(8,10))
plt.subplot(3,1,1)
plt.plot(df['Time [s]'].loc[:2000],df[' RESP'].loc[:2000])
plt.xlabel('Time [s]')
plt.title('Respiration level (Normalized)',fontsize=15)
plt.ylim(min(df[' RESP'].loc[:2000]),1.02)
plt.xlim(0,16)

plt.subplot(3,1,2)
plt.plot(df['Time [s]'].loc[:500],df[' PLETH'].loc[:500])
plt.xlabel('Time [s]')
plt.xlim(0,4)
plt.title('Plethysmogram',fontsize=15)

plt.subplot(3,1,3)
plt.plot(df['Time [s]'],df[' V'])
plt.xlabel('Time [s]')
plt.xlim(0,4)
plt.title('Voltage (ECG)',fontsize=15)
plt.subplots_adjust(hspace=0.5)
plt.show()
```
